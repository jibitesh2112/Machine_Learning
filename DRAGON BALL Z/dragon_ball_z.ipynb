```python
sns.set()
plt.figure(figsize=(15,8))
sns.heatmap(df.isnull(),cmap = 'viridis')
plt.show()
```

```python
# Organizing the number of songs released per year
charater_per_saga = (df.Saga_or_Movie.value_counts().rename_axis('Saga_or_Movie').reset_index(name = 'count'))
charater_per_saga
```

```python
# Plotting charater_per_saga in a bar chart
plt.figure(figsize=(14,8))
ax = sns.barplot(x = 'Saga_or_Movie', y = 'count', data = charater_per_saga,palette='deep')
plt.ylabel('Number of character', size = 15)
plt.xlabel('Number of saga or movies', size = 15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15, rotation = 90)
plt.show()
```

```python
# Organizing the number of songs released per year
charater_per_series = (df.Dragon_Ball_Series.value_counts().rename_axis('Dragon_Ball_Series').reset_index(name = 'count'))
charater_per_series
```

```python
# Plotting charater_per_saga in a bar chart
plt.figure(figsize=(14,8))
ax = sns.barplot(x = 'Dragon_Ball_Series', y = 'count', data = charater_per_series,palette='deep')
plt.ylabel('Number of character', size = 15)
plt.xlabel('Number of series', size = 15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15, rotation = 90)
plt.show()
```
